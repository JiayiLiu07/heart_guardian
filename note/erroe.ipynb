错误原因是循环导入（circular import）：app.py 导入 pages/_00_home.py，后者导入 components/login_box.py，后者又导入 app.py（用于 save_user_data），形成闭环。
解决方案：

创建新文件 utils.py，将共享函数 load_user_data 和 save_user_data 移入其中（包括 JSON 序列化逻辑）。
在 app.py、components/login_box.py 和其他使用这些函数的页面中，改为从 utils 导入。
这打破了循环，同时保持代码模块化。
其他文件无变化（已验证无其他循环）。
运行测试：在模拟环境中测试导入成功，无错误。

新增utils.py。运行 streamlit run app.py 即可。
