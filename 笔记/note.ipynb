原因：pyzbar 库在导入时无法找到底层的 libzbar 共享库。pyzbar 依赖于 libzbar 来处理条形码解码（如 QR 码或药品条形码扫描），而你的系统（macOS）缺少 libzbar 或其路径未正确配置。

TypeError: FormMixin.form_submit_button() got an unexpected keyword argument 'key'

这个错误表明，st.form_submit_button() 函数 不支持 key 参数。我在之前的修改中错误地给它添加了 key。

其次，一个潜在的 UX 问题：

Missing Submit Button 警告。虽然你使用了 st.form_submit_button(), 但 Streamlit 在某些情况下可能会发出这个警告，尤其是在表单内部的布局或逻辑处理不当的时候。不过，这个警告通常不会阻止功能，但值得注意。

解决方案：

移除 st.form_submit_button() 的 key 参数。
确保 st.form() 的 clear_on_submit 行为符合预期。
检查 st.session_state 的使用。
