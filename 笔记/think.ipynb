输入特征有 3 种类型：

目的：事实信息；
检查：体检结果；
主观：患者提供的信息。
特征：

年龄 | 客观特征 | 年龄 | int（天）
高度 | 目标特征 | 高度 | int (cm) |
重量| 目标特征| 重量| 浮点数（kg）|
性别 | 客观特征 | 性别 | 分类代码 |
收缩压 | 检查特征 | ap_hi | int |
舒张压 | 检查特征 | ap_lo | int |
胆固醇 | 检查特征 | 胆固醇 | 1：正常，2：高于正常，3：远高于正常 |
葡萄糖 | 检查特征 | 葡萄糖 | 1：正常，2：高于正常，3：远高于正常 |
吸烟 | 主观特征 | 烟雾 | 二进制 |
酒精摄入量 | 主观特征 | 酒精 | 二进制 |
